```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# copied from https://www.kaggle.com/jeffmoser/two-sigma-financial-modeling/kagglegym-api-overview/comments

with pd.HDFStore("train.h5", "r") as train:
    # Note that the "train" dataframe is the only dataframe in the file
    df = train.get("train")
```

```python
# grab all rows with id == 14. then graph timestamp of those ids vs y using seaborn library

id_10 = df[df['id'] == 14]
g = sns.jointplot("timestamp", "y", data=id_10, kind="reg")
```
